```python
import requests
from bs4 import BeautifulSoup

# URL of the page you want to scrape
url = 'https://economictimes.indiatimes.com/hdfc-bank-ltd/stocksupdate/companyid-9195.cms'

# Make an HTTP request to the URL
response = requests.get(url)
html_content = response.text

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find the elements containing the news headlines and dates
# Replace '.headline' and '.date' with the actual class names
articles = soup.find_all('div', class_='eachStory')  # Assuming each article is contained in this class
# Extract the headline and date text from each article element
data = []
for article in articles:
    headline = article.find(class_='headingText').text.strip()  # Replace with the actual class or tag
    date = article.find(class_='storyDate').text.strip()  # Replace with the actual class or tag
    data.append((date, headline))
```

```python
data[9][0].split(",")
```

```python
import pandas as pd
df = pd.DataFrame({"Date":[], "News":[]})

for x in data:
    time, title = x[0], x[1]
    temp = time.split(",")
    if len(temp) == 3:
        day = temp[0] + " " + temp[1]
    else:
        day = temp[0]
    headline = title.split("\n")[0]
    new_row = {"Date": day, "News": headline}
    df = df.append(new_row, ignore_index=True)
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Sample headlines
headlines = list(df["News"])

# Initialize the sentiment intensity analyzer
sia = SentimentIntensityAnalyzer()

# Function to categorize sentiment
def categorize_sentiment(score):
    if score > 0.05:
        return 'Positive'
    elif score < -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Analyzing sentiment of each headline
sentiments = []
new_list_senti = []
score_senti = []
for headline in headlines:
    score = sia.polarity_scores(headline)['compound']
    sentiment = categorize_sentiment(score)
    sentiments.append((headline, sentiment))
    new_list_senti.append(sentiment)
    score_senti.append(score)

# Display the results
for headline, sentiment in sentiments:
    print(f"Headline: {headline}\nSentiment: {sentiment}\n")

df["Score"] = score_senti
df["Sentiment"] = new_list_senti
```
