```python
import requests
from bs4 import BeautifulSoup

# URL of the page you want to scrape
url = 'https://economictimes.indiatimes.com/hdfc-bank-ltd/stocksupdate/companyid-9195.cms'

# Make an HTTP request to the URL
response = requests.get(url)
html_content = response.text

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find the elements containing the news headlines
# Replace '.headline' with the actual class name or HTML structure you observed
headlines = soup.find_all(class_='headingText')

print(headline)
# Extract the text from each headline
for headline in headlines:
    print(headline.text.strip())
```

```python
articles = soup.find_all('div', class_='eachStory')
# Assuming 'articles' is a ResultSet of article elements
for article in articles:
    # Use 'find' to get a single element from each article
    headline_element = article.find(class_='headingText')
    # Now you can safely check if the headline_element is not None and then extract text
    if headline_element:
        headline = headline_element.text.strip()
    else:
        headline = 'No headline found'

    # Similarly for the date
    date_element = article.find(class_='storyDate')
    if date_element:
        date = date_element.text.strip()
    else:
        date = 'No date found'

    # Now you can print or store the headline and date
    print(f'{date}: {headline}')
```
